```python
rng = np.random.default_rng(seed=42)
X = rng.random((10,2))
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

```python
dist_sq = np.sum((X[:,np.newaxis] - X[np.newaxis,:])**2, axis=-1)
```

```python
difference = X[:,np.newaxis] - X[np.newaxis,:]
difference.shape
```

```python
difference
```

```python
sqDiff = difference**2
sqDiff.shape
```

```python
sumSqDiff=np.sum(sqDiff,axis=-1)
```

```python
sumSqDiff.shape
```

```python
sumSqDiff
```

```python
sumSqDiff.diagonal()
```
