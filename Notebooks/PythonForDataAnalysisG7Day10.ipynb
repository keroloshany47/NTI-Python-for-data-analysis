```python
x = np.linspace(0,10,100)

plt.plot(x,np.sin(x),color='blue')
plt.plot(x,np.sin(x-1),color='g')#rgbcmyk
plt.plot(x,np.sin(x-2),color='0.75')
plt.plot(x,np.sin(x-3),color='#FFDD44')#hex code (RRGGBB, 00 to FF)
plt.plot(x, np.sin(x-4), color=(1.0,.2,.3))#RGB tuple
plt.plot(x,np.sin(x-5),color='chartreuse')
```

```python
plt.plot(x,x,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted')
```

```python
plt.plot(x,x,linestyle='-')
plt.plot(x,x+1,linestyle='--')
plt.plot(x,x+2,linestyle='-.')
plt.plot(x,x+3,linestyle=':')
```

```python
plt.plot(x,x,'-g',label='y=x')
plt.plot(x,x+1,'--c',label='y=x+1')
plt.plot(x,x+2,'-.k',label='y=x+2')
plt.plot(x,x+3,':r',label='y=x+3')
plt.legend()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.cos(x))
ax.set_xlabel('x')
ax.set_ylabel('cos(x)')
ax.set_xlim(0,10)
ax.set_ylim(-1,1)
ax.set_title('A cosine curve')
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.cos(x))
ax.set(xlim=(0,10),ylim=(-1,1),title='cosine curve')
ax.grid()
```
